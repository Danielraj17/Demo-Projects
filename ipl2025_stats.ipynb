Visualize the Data's

```python
data['toss_decision'].value_counts().plot.pie(y = 'toss_decision',figsize=(5,5),)
plt.title("The Captain Choise After Toss ")
```

```python
top_teams = data['match_winner'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_teams.values, y=top_teams.index, palette='crest')
plt.title('Teams by Number of Wins')
plt.xlabel('Number of Wins')
plt.ylabel('Teams')
plt.show()
```

```python
sns.countplot(x="toss_decision",data=data)
plt.title("Toss Decision Counts")
plt.show()
```

```python
venue_counts = data['venue'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette='magma')
plt.title(' Venues by Number of Matches')
plt.xlabel('Number of Matches')
plt.show()
```

```python
team_matches = pd.concat([data['team1'], data['team2']]).value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=team_matches.values, y=team_matches.index, palette='coolwarm')
plt.title('Most Matches Played by Teams')
plt.xlabel('Number of Matches')
plt.ylabel('Teams')
plt.show()
```

```python
sns.barplot(x=top_players.values, y=top_players.index)
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()
```
