```python
df = data.fillna('-')
df.isna().sum()
```

```python
year_cols = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

total_crimes_per_year = data[year_cols].sum()

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(x=total_crimes_per_year.index, y=total_crimes_per_year.values, marker='o', linewidth=2)

plt.title("Total Crimes in Tamil Nadu Over the Years (20142022)")
plt.xlabel("Year")
plt.ylabel("Total Crimes")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_cleaned = df.dropna(subset=['DISTRICT', 'Total _Crimes'])

df_sorted = df_cleaned.sort_values(by='Total _Crimes', ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='Total _Crimes', y='DISTRICT', data=df_sorted, palette='viridis')

plt.title("Crime Distribution by District (20142022 Total)")
plt.xlabel("Total Crimes")
plt.ylabel("District")
plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
df_trends = df.dropna(subset=['DISTRICT'])  

year_cols = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']


df_trends[year_cols] = df_trends[year_cols].fillna(0)

# Plot for top 5 districts with highest total crimes
top_districts = df_trends.sort_values(by='Total _Crimes', ascending=False).head(5)

plt.figure(figsize=(12, 7))

# Plot each district
for _, row in top_districts.iterrows():
    plt.plot(year_cols, row[year_cols], marker='o', label=row['DISTRICT'])

plt.title("District-wise Crime Trends (Top 5 Districts)")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.legend(title="District")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

north_districts = [
    'Chennai', 'Kanchipuram', 'Tiruvallur', 'Vellore', 'Tiruvannamalai',
    'Cuddalore', 'Villupuram', 'Nagapattinam', 'Perambalur', 'Ariyalur',
    'Thiruvarur'
]

south_districts = [
    'Madurai', 'Thoothukudi', 'Tirunelveli', 'Dindigul', 'Virudhunagar',
    'Ramanathapuram', 'Sivagangai', 'Theni', 'Pudukottai', 'Tenkasi'
]

df_region = data.dropna(subset=['DISTRICT', 'Total _Crimes']).copy()

df_region['Region'] = df_region['DISTRICT'].apply(
    lambda x: 'North' if x in north_districts else ('South' if x in south_districts else 'Other')
)

df_filtered = df_region[df_region['Region'].isin(['North', 'South'])]

region_crime = df_filtered.groupby('Region')['Total _Crimes'].sum().reset_index()

print(region_crime)

# Barploot 

plt.figure(figsize=(8, 5))
sns.barplot(data=region_crime, x='Region', y='Total _Crimes', palette='Set2')

plt.title("Total Crimes: North vs South Tamil Nadu")
plt.ylabel("Total Crimes (20142022)")
plt.xlabel("Region")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
